```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import fourier_series, pi, plot
from sympy.abc import x,L,H
from sympy import Piecewise
import sympy as sp
```

```python
# global constants

n = 256
```

```python
### given function of x ###

def f(x):
    
    ff = -999.0*np.ones(len(x))
    
    for i in range(0,len(x)):
        if(x[i] < 0.6):
            ff[i] = 2.0*x[i]
        else:
            ff[i] = 0.0
    return ff
```

```python
### integral of f(x) ###

def intf(xx):

    if(xx < 0.6):
        nf = xx*xx
    else:
        nf = 0.36
    return nf
```

```python
### the haar wavelets ###

def haar(x,a,b):
    
    h = np.zeros(n)
    
    if(a==0):
        for i in range(0,n):
            h[i] = 1
            
    if(a==1):
        for i in range(0,int(n/2)):
            h[i] = 1       
        for i in range(int(n/2),n):
            h[i] = -1
            
    if(a==2):
        for i in range(0, int(n / 4)):
            h[i] = 1       
        for i in range(int(1 * n / 4),int(2 * n / 4)):
            h[i] = -1
        for i in range(int(2 * n / 4),int(3 * n / 4)):
            h[i] = 1
        for i in range(int(3 * n / 4),int(4 * n / 4)):
            h[i] = -1
            
    if(a==3):
        for i in range(0, int(1 * n / 8)):
            h[i] = 1       
        for i in range(int(1 * n / 8),int(2 * n / 8)):
            h[i] = -1
        for i in range(int(2 * n / 8),int(3 * n / 8)):
            h[i] = 1
        for i in range(int(3 * n / 8),int(4 * n / 8)):
            h[i] = -1
        for i in range(int(4 * n / 8),int(5 * n / 8)):
            h[i] = 1
        for i in range(int(5 * n / 8),int(6 * n / 8)):
            h[i] = -1
        for i in range(int(6 * n / 8),int(7 * n / 8)):
            h[i] = 1
        for i in range(int(7 * n / 8),int(8 * n / 8)):
            h[i] = -1
            
    return h
           
```

```python
### the coefficients ###

def coeffs(a,b):
    
    h = np.zeros(n)
    
    if(a==0):
        cc = (intf(1.00) - intf(0.00))

    if(a==1):
        cc = (intf(1/2)-intf(0.00)) - (intf(1.00)-intf(1/2))
        
    if(a==2):
        cc = - ((intf(2/4)-intf(1/4)) - (intf(1/4)-intf(0/4)) - (intf(1.00)-intf(3/4) - (intf(3/4)-intf(2/4))))
        
    if(a==3):
        cc = ((intf(8/8)-intf(7/8)) - (intf(7/8)-intf(6/8)) - (intf(6/8)-intf(5/8)) - (intf(5/8)-intf(4/8)) - (intf(4/8)-intf(3/8)) - (intf(3/8)-intf(2/8)) - (intf(2/8)-intf(1/8)) - (intf(1/8)-intf(0/8)))
            
    return cc
  
```

```python
### main script ###

x  = np.linspace(0.0, 1.0, n, endpoint=False)
ff = f(x)
# nf = intf(x)

plt.plot(x,ff)
# plt.plot(x,nf)
plt.show()
```

```python
# Creating the different Haar Wavelets

# a = 0, b = 0
h00 = haar(x, 0, 0)
c00 = coeffs(0, 0)
print("C00 =",c00)
print(f'h00 = {h00}')

plt.title("$C_{00} h_{00}$")
plt.plot(x,ff,label="$f(x)$")
plt.plot(x,c00*h00,label="$C_{00} h_{00}$")
plt.legend()
plt.show()


# a = 1, b = 0
h10 = haar(x, 1, 0)
c10 = coeffs(1, 0)
print("C10 =",c10)
print(f'h10 = {h10}')
# for i in range(len(h10)):
#     print(i,h10[i])

plt.title('$C_{00} h_{00} + C_{10} h_{10}$')
plt.plot(x,ff,label="$f(x)$")
plt.plot(x,c00*h00 + c10*h10,label="$C_{00} h_{00} + C_{10} h_{10}$")
plt.legend()
plt.show()


# a = 2, b = 0
h20 = haar(x, 2, 0)
c20 = coeffs(2, 0)
print("C20 =",c20)
print(f'h20 = {h20}')
# for i in range(len(h20)):
#     print(i,h20[i])

plt.title('$C_{00} h_{00} + C_{10} h_{10} + C_{20} h_{20}$')
plt.plot(x, ff, label="$f(x)$")
plt.plot(x, c00*h00 + c10*h10 + c20*h20, label="$C_{00} h_{00} + C_{10} h_{10} + C_{20} h_{20}$")
plt.legend()
plt.show()

# a = 3, b = 0
h30 = haar(x, 3, 0)
c30 = coeffs(3, 0)
print("C30 =",c30)
print(f'h30 = {h30}')
# for i in range(len(h30)):
#     print(i,h30[i])

plt.title('$C_{00} h_{00} + C_{10} h_{10} + C_{20} h_{20} + C_{30} h_{30}$')
plt.plot(x, ff, label="$f(x)$")
plt.plot(x, c00*h00 + c10*h10 + c20*h20 + c30*h30, label="$C_{00} h_{00} + C_{10} h_{10} + C_{20} h_{20} + C_{30} h_{30}$")
plt.legend()
plt.show()
```
